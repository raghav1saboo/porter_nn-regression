```python
from sklearn.model_selection import train_test_split
from category_encoders import TargetEncoder as te

x_train, x_test, y_train, y_test = train_test_split(df_1.drop(columns='delivery_time'), df_1['delivery_time'], test_size=0.2, random_state=45)
te_enc = te(smoothing=10,return_df=True)
x_train = te_enc.fit_transform(x_train, y_train)
x_test = te_enc.transform(x_test)

```

```python
from sklearn.preprocessing import StandardScaler as ss
from sklearn.ensemble import IsolationForest

scaler = ss()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

x_train_scaled = pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled = pd.DataFrame(x_test_scaled,columns=x_test.columns)

iso = IsolationForest(n_estimators=234,contamination='auto',max_samples=1024,bootstrap=True,n_jobs=-1,)
x_train_scaled['outlier'] = iso.fit_predict(x_train_scaled)
x_test_scaled['outlier'] = iso.predict(x_test_scaled)

print(x_train_scaled['outlier'].value_counts())
print(x_test_scaled['outlier'].value_counts())
x_train_scaled = x_train_scaled.loc[x_train_scaled['outlier']==1].drop(columns='outlier')
y_train = y_train.iloc[x_train_scaled.index]

x_test_scaled = x_test_scaled.loc[x_test_scaled['outlier']==1].drop(columns='outlier')
y_test = y_test.iloc[x_test_scaled.index]
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import layers
from keras.optimizers import Adam
from keras.losses import MeanSquaredError
from keras.metrics import MeanAbsoluteError
import keras_tuner as kt

model = keras.Sequential()
model.add(layers.Input(shape=(x_train_scaled.shape[1],)))
model.add(layers.Dense(units=9,activation='leaky_relu'))
model.add(layers.Dense(units=18,activation='leaky_relu'))
model.add(layers.Dense(units=36,activation='leaky_relu'))
model.add(layers.Dense(units=72,activation='leaky_relu'))

model.add(layers.Dense(1))
model.compile(optimizer=Adam(learning_rate=0.001),loss=MeanSquaredError(),metrics=[MeanAbsoluteError()])
print("Fit model on training data")
model.fit(x_train_scaled,y_train,epochs=45,validation_split=0.2,verbose=1,batch_size=1024)
print("Evaluate on test data")
results = model.evaluate(x_test_scaled, y_test, batch_size=1024)
print("test loss, test acc:", results)
# model.summary()
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import layers
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.model_selection import KFold
import numpy as np
import kerastuner as kt
from keras.initializers import HeUniform

# Define the model builder function for Keras Tuner
def model_builder(hp):
    model = keras.Sequential()

    # Input layer (explicitly define input shape if known)
    model.add(keras.Input(shape=(x_train_scaled.shape[1],)))  # Replace input_shape

    # Tune the number of hidden layers
    for i in range(hp.Int('num_layers', min_value=1, max_value=9)):
        # Tune the number of units in each layer
        units = hp.Int(f'units_{i}', min_value=18, max_value=90, step=36)

        # Tune the activation function
        activation = hp.Choice(f'activation_{i}', values=['leaky_relu', 'tanh', 'elu', 'softplus', 'softsign', 'selu', 'exponential', 'silu','gelu','mish'])
        # Tune regularization (L1, L2, or None)
        regularizer = None
        reg_type = hp.Choice(f'regularizer_{i}', values=['l1', 'l2', 'none'])
        if reg_type == 'l1':
            reg_strength = hp.Float(f'reg_strength_{i}', min_value=1e-5, max_value=1e+2, sampling='log')
            regularizer = keras.regularizers.L1(reg_strength)
        elif reg_type == 'l2':
            reg_strength = hp.Float(f'reg_strength_{i}', min_value=1e-5, max_value=1e+2, sampling='log')
            regularizer = keras.regularizers.L2(reg_strength)

        # model.add(layers.Dense(units=units, activation=activation, kernel_regularizer=regularizer))

        model.add(layers.Dense(units=units, activation=activation,kernel_initializer=HeUniform(),bias_initializer="zeros",kernel_regularizer=regularizer,activity_regularizer=regularizer))

        # Tune dropout
        if hp.Boolean(f'dropout_{i}'):
            dropout_rate = hp.Float(f'dropout_rate_{i}', min_value=0.1, max_value=0.5, step=0.1)
            model.add(layers.Dropout(dropout_rate))

        # Tune batch normalization
        if hp.Boolean(f'batchnorm_{i}'):
            model.add(layers.BatchNormalization())


    # Output layer (linear activation for regression)
    model.add(layers.Dense(1, activation='linear'))

    # Tune the learning rate
    lr = hp.Float('learning_rate', min_value=1e-4, max_value=1e+2, sampling='log')
    optimizer = keras.optimizers.Adam(learning_rate=lr) #You can change to other optimizers
    model.compile(optimizer=optimizer, loss='mse', metrics=['mae']) # You can change the loss function

    return model

# Sample data (replace with your actual data)
# num_samples = 1000
# input_shape = 10
# X = np.random.rand(num_samples, input_shape)
# y = np.random.rand(num_samples, 1)

# # K-fold cross-validation
kfold = KFold(n_splits=5, shuffle=True, random_state=45)

best_models = []
fold_no = 1
for train, test in kfold.split(x_train_scaled, y_train):
    print(f"Fold {fold_no}")

    # Instantiate the tuner (BayesianOptimization is a good starting point)
    tuner = kt.BayesianOptimization(
        model_builder,
        objective='val_loss',
        max_trials=9,  # Adjust the number of trials
        executions_per_trial=3,
        directory='my_dir',
        project_name=f'regression_tuning_fold_{fold_no}'
    )

    # Callbacks
    stop_early = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6)
    callbacks = [stop_early, reduce_lr]

    tuner.search(x_train_scaled.iloc[train], y_train.iloc[train], epochs=45, validation_data=(x_train_scaled.iloc[test], y_train.iloc[test]), callbacks=callbacks, verbose=1) #You can change the number of epochs

    # Get the best model from the current fold
    best_model = tuner.get_best_models(num_models=1)[0]
    best_models.append(best_model)
    fold_no += 1

# Evaluate the best models from each fold (optional)
for i, model in enumerate(best_models):
    loss, mae = model.evaluate(x_train_scaled, y_train, verbose=1)
    print(f"Fold {i+1} Best Model - Loss: {loss:.4f}, MAE: {mae:.4f}")

# You can now choose the best model among the folds or average their predictions.
# Or retrain the model on the whole dataset using the combined hyperparameters.
```
